```python
import pandas as pd
import requests
from matplotlib import pyplot as plt
import json
from pprint import pprint
```

```python
outbreak_df = pd.read_csv("NationalOutbreakPublicDataTool.csv")


outbreak_df.dropna(subset=["Etiology"], inplace = True)
outbreak_df.columns

clean_outbreak_df = outbreak_df[['Year', 'Month', 'State', 'Primary Mode', 'Etiology','Etiology Status', 'Setting','Illnesses',
       'Hospitalizations', 'Info on Hospitalizations', 'Deaths',
       'Info on Deaths']]
clean_outbreak_df.count()
```

```python
clean_outbreak_df.dropna(subset=["Hospitalizations"], inplace=True)

clean_outbreak_df.count()
```

```python
clean_outbreak_df.dropna(subset=["Deaths"], inplace=True)
clean_outbreak_df.count()
```

```python
clean_outbreak_df.dropna(subset=["Setting"], inplace=True)
clean_outbreak_df.count()
```

```python
final_outbreak_df = clean_outbreak_df
final_outbreak_df
```

```python
final_outbreak_df.to_csv("Cleaned_outbreak_data.csv", index=False)
```
